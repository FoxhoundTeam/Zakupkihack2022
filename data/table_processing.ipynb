# [+] Выделение характеристик/артибутов простое строковое

1. Сгруппировать по типу
2. Собрать все пары свойство: значение для каждой группы рег. выражениями
2. Построить словарь свойств
3. Выделить множество значений для каждого свойства

```python
string = "||".join(df2.groupby(["okpd2_code"])["product_characteristics"].apply(list)[2])
import re
reg_expr = "(^\s*([^\|:]+)\s*:\s*([^\|:]+)\s*\|)|(\|\s*([^\|:]+)\s*:\s*([^\|:]+)\s*\|)|(\|\s*([^\|:]+)\s*:\s*([^\|:]+)\s*$)"
re.findall(reg_expr, string)

matches = re.findall(reg_expr, string)
filters = {}
for itm in matches:
    key = itm[1].lower().strip()
    val = itm[2].lower().strip()
    if key not in filters:
        filters[key] = []
    if val not in filters[key]:
        filters[key].append(val)
    
    key = itm[4].lower().strip()
    val = itm[5].lower().strip()
    if key not in filters:
        filters[key] = []
    if val not in filters[key]:
        filters[key].append(val)
    
    key = itm[7].lower().strip()
    val = itm[8].lower().strip()
    if key not in filters:
        filters[key] = []
    if val not in filters[key]:
        filters[key].append(val)

del filters['']
del filters['описание']
filters
```

```python
groupby_series = df2.groupby(["okpd2_name"])["product_characteristics"].apply(list)
type(groupby_series)
# df_filters = pd.DataFrame(groupby_series)
# df_filters
result = []
result_columns = ["okpd2_code","filter_name","variants"]
reg_expr = r"(^\s*([^\|:]+)\s*:\s*([^\|:]+)\s*\|)|(\|\s*([^\|:]+)\s*:\s*([^\|:]+)\s*\|)|(\|\s*([^\|:]+)\s*:\s*([^\|:]+)\s*$)"

import re

for index, value in groupby_series.items():
    # print(index)
    # print(value)
    characteristics = "||".join(value)
    matches = re.findall(reg_expr, characteristics)
    filters = {}
    for itm in matches:
        key = itm[1].lower().strip()
        val = itm[2].lower().strip()
        if key not in filters:
            filters[key] = []
        if val not in filters[key]:
            filters[key].append(val)

        key = itm[4].lower().strip()
        val = itm[5].lower().strip()
        if key not in filters:
            filters[key] = []
        if val not in filters[key]:
            filters[key].append(val)

        key = itm[7].lower().strip()
        val = itm[8].lower().strip()
        if key not in filters:
            filters[key] = []
        if val not in filters[key]:
            filters[key].append(val)
    if '' in filters:
        del filters['']
    if ' ' in filters:
        del filters[' ']
    if 'описание' in filters:
        del filters['описание']
    for filter_name in filters:
        result.append([index, filter_name, filters[filter_name]])

# result
```

```python
result_df = pd.DataFrame(result, columns=result_columns)
result_df.to_csv("filters.csv", sep=";")
result_df.head(5)
```
